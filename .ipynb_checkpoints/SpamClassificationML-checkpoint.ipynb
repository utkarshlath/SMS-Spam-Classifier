# SMS SPAM CLASSIFICATION

## Import necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.corpus import stopwords
import string
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib
import warnings
warnings.filterwarnings('ignore')
```

## Data Loading & Cleaning

```python
#Loading the dataset
messages = pd.read_csv("./data/SpamCollection.csv", encoding='latin-1')
messages.head(5)
```

```python
#Removing unnecessary columns and changing column names
messages = messages.iloc[:,[0,1]]
messages.columns = ["label", "message"]
messages.head()
```

```python
#Describing our dataset
messages.describe()
```

##  Data Visualization & Feature Engineering

```python
#Finding the composition of our dataset
messages["label"].value_counts().plot(kind = 'pie', explode = [0, 0.1], figsize = (5, 5), autopct = '%1.1f%%', shadow = True)
plt.ylabel("Spam vs Ham")
plt.legend(["Ham", "Spam"])
plt.show()
```

```python
#Most common messages in each category
messages.groupby('label').describe()
```

```python
# Creating new feature messages length
messages['length'] = messages['message'].apply(len)
messages.head()
```

```python
# Checking out the newly created feature
messages.length.describe()
```

```python
# Plot to find out frequency of length of sms messages
plt.style.use('fivethirtyeight')
fig = plt.figure(dpi=100)
ax = plt.axes()
messages['length'].plot(bins=50, kind='hist',ax=ax)
ax.set(xlabel = 'Message Length',ylabel = 'Frequency',title = 'Length');
```

```python
# Plots to find out length distribution among spam vs non-spam messages
figure,axis = plt.subplots(1,2,figsize=(12,4),dpi = 120)



sns.distplot(messages[messages['label']=='ham']["length"], bins = 50, ax = axis[0], color = 'green')
axis[0].set_xlabel("Message Length")
axis[0].set_ylabel("Frequency")
axis[0].set_title("NON-SPAM MESSAGES")

sns.distplot(messages[messages["label"] == "spam"]["length"], bins = 50, ax = axis[1], color="red")
axis[1].set_xlabel("Message Length")
axis[1].set_ylabel("Frequency")
axis[1].set_title("SPAM MESSAGES")
```

```python
# Extracting individual spam and ham messages
spam_messages = messages[messages["label"] == "spam"]["message"]
ham_messages = messages[messages["label"] == "ham"]["message"]
```

```python
# Find out words from the messages
def extractWords(messages):
    extracted_words = []
    words = [word.lower() for word in word_tokenize(messages) if word.lower() not in stopwords.words("english") and word.lower().isalpha()]
    extracted_words = extracted_words + words
    return extracted_words
```

```python
spam_words = spam_messages.apply(extractWords)
ham_words = ham_messages.apply(extractWords)
```

```python
# Create Word Cloud for SPAM messages
wordcloudspam = WordCloud(width=600, height=400, colormap="Spectral").generate((' '.join([str(elem) for elem in spam_words])))
plt.figure(figsize=(10,8))
plt.imshow(wordcloudspam)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Create Word Cloud for NON-SPAM messages
wordcloudham = WordCloud(width=600, height=400, colormap="Spectral").generate((' '.join([str(elem) for elem in ham_words])))
plt.figure(figsize=(10,8))
plt.imshow(wordcloudham)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

## Creating models and training various classifiers

```python
sms_data = messages["message"].copy()
```

```python
def processing_messages(textData):
    textData = textData.translate(str.maketrans('', '', string.punctuation))
    textData = [word for word in textData.split() if word.lower() not in stopwords.words('english')]
    return " ".join(textData)
```

```python
sms_data = sms_data.apply(processing_messages)
tfid_vectorizer = TfidfVectorizer(stop_words="english")
tfid_vectorizer_transformed = tfid_vectorizer.fit_transform(sms_data)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(tfid_vectorizer_transformed, messages['label'], test_size=0.3, random_state=111)
```

### Hyperparameter tuning for KNN

```python
plt.style.use('seaborn')
pred_scores = []
for i in range(5,51):
    kneighbors_classifier = KNeighborsClassifier(n_neighbors=i)
    kneighbors_classifier.fit(X_train, Y_train)
    prediction = kneighbors_classifier.predict(X_test)
    pred_scores.append((i, accuracy_score(Y_test,prediction)*100))

scores = pd.DataFrame(pred_scores,columns=['K Value','Score'])
plt.plot(scores['K Value'], scores['Score'])
plt.xlabel("K Values")
plt.ylabel("Scores")
plt.title("Accuracy for various values of k")
plt.show()
```

```python
scores[scores['Score'] == scores['Score'].max()]
```

### Hyperparameter tuning for Multinomial NB

```python
pred_scores = []
for i in np.linspace(0.05, 1, num=20):
    multinomialnb_classifier = MultinomialNB(alpha=i)
    multinomialnb_classifier.fit(X_train, Y_train)
    prediction = multinomialnb_classifier.predict(X_test)
    pred_scores.append((i, accuracy_score(Y_test,prediction)*100))
scores = pd.DataFrame(pred_scores,columns=['Alpha Value','Score'])
plt.plot(scores['Alpha Value'], scores['Score'])
plt.xlabel("Alpha Values")
plt.ylabel("Scores")
plt.title("Accuracy for various values of alpha")
plt.show()
```

```python
scores[scores['Score'] == scores['Score'].max()]
```

### Hyperparameter tuning for Random Forest

```python
pred_scores = []
for i in range(2,51):
    randomforest_classifier = RandomForestClassifier(n_estimators=i, random_state=100)
    randomforest_classifier.fit(X_train, Y_train)
    prediction = randomforest_classifier.predict(X_test)
    pred_scores.append((i, accuracy_score(Y_test,prediction)*100))
scores = pd.DataFrame(pred_scores,columns=['Estimators','Score'])
plt.plot(scores['Estimators'], scores['Score'])
plt.xlabel("No. of Estimators")
plt.ylabel("Scores")
plt.title("Accuracy for different estimator values")
plt.show()
```

```python
scores[scores['Score'] == scores['Score'].max()]
```

### Using the hyperparameters values found to train and test models

```python
#Defining models
kneighbors_classifier = KNeighborsClassifier(n_neighbors=49)
multinomialnb_classifier = MultinomialNB(alpha=0.25)
randomforest_classifier = RandomForestClassifier(n_estimators=33, random_state=100)
```

```python
def train(classifier, X_train, Y_train):    
    classifier.fit(X_train, Y_train)
```

```python
def predict(classifier, X_test):
    return (classifier.predict(X_test))
```

```python
classifiers = {"K Nearest Neighbors": kneighbors_classifier, "Multinomial Naive Bayes": multinomialnb_classifier, "Random Forest": randomforest_classifier}
pred_scores = []
for key,value in classifiers.items():
    train(value, X_train, Y_train)
    prediction = predict(value,X_test)
    pred_scores.append((key, accuracy_score(Y_test,prediction)*100))
```

```python
scores = pd.DataFrame(pred_scores,columns=['Classifier','Accuracy'])
scores
```

```python
scores.plot(kind='bar',  ylim=(85,100), figsize=(5,5), colormap="Set2")
plt.xticks(np.arange(3), scores["Classifier"], rotation=20)
plt.ylabel('Accuracy Score')
plt.title('Accuracy by Classifier')
```

### Using stemming

```python
def stemmer(text):
    text = text.split()
    words = ""
    for i in text:
            stemmer = SnowballStemmer("english")
            words += (stemmer.stem(i))+" "
    return words
```

```python
sms_data = sms_data.apply(stemmer)
tfid_vectorizer_transformed = tfid_vectorizer.fit_transform(sms_data)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(tfid_vectorizer_transformed, messages['label'], test_size=0.3, random_state=111)
```

```python
pred_scores = []
for key,value in classifiers.items():
    train(value, X_train, Y_train)
    prediction = predict(value,X_test)
    pred_scores.append((key, accuracy_score(Y_test,prediction)*100))
```

```python
scores1 = pd.DataFrame(pred_scores,columns=['Classifier','Accuracy1'])
```

```python
scoresmerged = pd.concat([scores.set_index("Classifier"),scores1.set_index("Classifier")],axis=1)
scoresmerged
```

```python
scoresmerged.plot(kind='bar',  ylim=(85,100), figsize=(6,7), colormap="Set2")
plt.xticks(np.arange(3), scores["Classifier"], rotation=20)
plt.ylabel('Accuracy Score')
plt.title('Accuracy by Classifier')
```

### Using Length as a feature

```python
lf = messages['length'].to_numpy()
newsms = np.hstack((tfid_vectorizer_transformed.todense(),lf[:, None]))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(newsms, messages['label'], test_size=0.3, random_state=111)
```

```python
pred_scores = []
for key,value in classifiers.items():
    train(value, X_train, Y_train)
    prediction = predict(value,X_test)
    pred_scores.append((key, accuracy_score(Y_test,prediction)*100))
```

```python
scores2 = pd.DataFrame(pred_scores,columns=['Classifier','Accuracy2'])
```

```python
allscores = pd.concat([scoresmerged,scores2.set_index("Classifier")],axis=1)
allscores
```

## Combined accuracy of all three models in all three cases

```python
allscores.plot(kind='bar',  ylim=(85,100), figsize=(6,7), colormap="Set2")
plt.xticks(np.arange(3), scores["Classifier"], rotation=20)
plt.ylabel('Accuracy Score')
plt.title('Accuracy by Classifier')
```
